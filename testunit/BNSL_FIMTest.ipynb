```python
import pyAgrum.lib.bn_vs_bn as gcm
import pyAgrum.lib.notebook as gnb
def dict2html(di1,di2=None): 
    res= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di1.items()])
    if di2 is not None:
        res+="<br/><br/>"
        res+= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di2.items()])
    return res

bn = asiatest.ori_bn()
bnsl = asiatest.bnsl_bn()
bdeu = asiatest.bdeu_bn()

cmp=gcm.GraphicalBNComparator(bnsl,bn)
#kl=gum.ExactBNdistance(bn_o,bn)
gnb.sideBySide(bnsl,bn,gnb.getBNDiff(bnsl,bn),dict2html(cmp.scores(),cmp.hamming()),cmp.equivalentBNs(),
              captions=['bn1','bn2','graphical diff','Scores','equivalent ?','distances'])
```
